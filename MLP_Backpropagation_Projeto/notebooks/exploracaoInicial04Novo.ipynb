# Experimento: Influência da Quantidade e Qualidade dos Dados em Redes Neurais

Este experimento avalia como diferentes **quantidades** e **qualidades** (dados com ou sem ruído) impactam o desempenho de uma rede neural simples no dataset Fashion MNIST.

## 1. Importações

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.initializers import GlorotUniform
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import MeanSquaredError
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

## 2. Carregamento e Normalização dos Dados

```python
# Carregar dados e normalizar
(x_train, y_train), (_, _) = tf.keras.datasets.fashion_mnist.load_data()
x_train = x_train / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
```

## 3. Função para Adicionar Ruído

```python
# Função para adicionar ruído gaussiano (opcional)
def adicionar_ruido(X, ruido_std=0.1):
    ruido = np.random.normal(loc=0.0, scale=ruido_std, size=X.shape)
    X_noisy = X + ruido
    X_noisy = np.clip(X_noisy, 0., 1.)  # garantir que fique entre 0 e 1
    return X_noisy
```

## 4. Parâmetros do Experimento

```python
# Tamanhos diferentes para testar
tamanhos = [1000, 5000, 10000, 30000]

# Qualidade dos dados: 'limpo' ou 'ruidoso'
qualidades = ['limpo', 'ruidoso']
```

## 5. Definição do Modelo

```python
def build_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(64, activation='relu', kernel_initializer=GlorotUniform(seed=0)),
        Dense(10, activation='softmax')
    ])
    return model
```

## 6. Execução dos Testes e Plotagem dos Resultados

```python
loss_fn = MeanSquaredError()
results = {}

for qualidade in qualidades:
    print(f"\n### Qualidade dos dados: {qualidade} ###")
    for n in tamanhos:
        print(f"Treinando com {n} exemplos")

        x_subset = x_train[:n]
        y_subset = y_train[:n]

        if qualidade == 'ruidoso':
            x_subset = adicionar_ruido(x_subset, ruido_std=0.1)

        model = build_model()
        model.compile(optimizer=SGD(learning_rate=0.01),
                      loss=loss_fn,
                      metrics=['mse'])

        history = model.fit(x_subset, y_subset, epochs=20, verbose=0)
        loss_curve = history.history['loss']

        results[(qualidade, n)] = loss_curve
        plt.plot(loss_curve, label=f'{qualidade} - {n} exemplos')

plt.title('Curvas de perda (Loss) – Quantidade e Qualidade dos Dados')
plt.xlabel('Época')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()
```

## 7. Avaliação com Medida-F (F1 Score)

Além da curva de perda, utilizamos a Medida-F para avaliar a performance do modelo com diferentes quantidades e qualidades de dados. Essa métrica é especialmente útil em problemas de classificação multiclasse, pois leva em consideração tanto a precisão quanto o recall.

A seguir, dividimos os dados em treino e teste (80/20), treinamos o modelo e calculamos a Medida-F usando `f1_score` do `sklearn.metrics`.


```python
x_tr, x_ts, y_tr, y_ts = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

model = build_model()
model.compile(optimizer=SGD(learning_rate=0.01),
              loss=loss_fn,
              metrics=['accuracy'])

model.fit(x_tr, y_tr, epochs=20, verbose=0)

y_pred = model.predict(x_ts)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_ts, axis=1)

f1 = f1_score(y_true, y_pred_classes, average='macro')
print(f"🟢 Medida-F (macro): {f1:.4f}")
```
